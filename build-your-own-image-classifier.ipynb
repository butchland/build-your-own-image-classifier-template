```python
#hide
from fastai.vision.all import *
from ipywidgets import widgets 
from fastai.vision.widgets import *
```

```python
#hide
path = Path()
# path.ls(file_exts='.pkl')
```

```python
learn_inf = load_learner(path/'export.pkl')
```

# My Very Own (Really Cool!)  Image Classifier

Classifier Categories:

```python
for category in learn_inf.dls.vocab:
    print(category)
```

```python
btn_upload = SimpleNamespace(data = ['images/purple_dog.jpg'])
```

```python
img = PILImage.create(btn_upload.data[-1])
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(200,200))
# out_pl
```

```python
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
# lbl_pred
```

```python
btn_run = widgets.Button(description='Classify')
# btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
#hide
#Putting back btn_upload to a widget for next cell
btn_upload = widgets.FileUpload()
```

```python
VBox([widgets.Label('UPLOAD AND CLASSIFY AN IMAGE!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
